###### Find jobs according to experience and salary expectation

```python
def myfilter(x, exp):
    if type(x) == str and "yrs" in x and "-" in x:
        s = str(x).split("-")
        a = int(s[0])
        b = int(s[1][:s[1].index("y")])
        if exp in range(a, b):
            return True
        elif exp==0 and exp==a:
            return True
        return False
    if str(exp) in x:
        return True
    return False
```

```python
def expfilter(x, exp):
    if type(x) == str and "yrs" in x and "-" in x:
        s = str(x).split("-")
        a = int(s[0])
        b = int(s[1][:s[1].index("y")])
        if exp in range(a, b):
            return True
        elif exp==0 and exp==a:
            return True
        return False
    if str(exp) in x:
        return True
    return False


def salfilter(x,sal):
    if type(x) == str and "-" in x and len(x)<=26 and "pa." in x.lower():# if cr crore lkh lakh la
        x = x[:x.index("PA")+2]
        s=str(x).split('-')
        p=s[0].replace(',','').strip()
        q=s[1].replace(',','').strip()
        
        try:
            a=int(p)
            b=int(q[:q.index("PA")])
        except:
            return False
        else:
            if sal in range(a,b):
                return True
            else:
                return False
    return False


exp = input("\n Enter your experience in numbers : ")
if exp.isdigit():
    sal = (input("\n Enter the salary expection PA : "))
    if sal.isdigit():
        temp_=data[data['Job Experience Required'].apply(lambda x: myfilter(str(x), int(exp)))]
        print(temp_[temp_['Job Salary'].apply(lambda x: salfilter(str(x),int(sal)))][['Job Experience Required',"Job Salary"]])
    else:
        print("\n SALARY SHOULD BE IN NUMBERS")
else:
    print("\n EXPERIENCE SHOULD BE IN NUMBERS")

```

```python
def colors():
    import numpy as np
    return "#"+"".join(np.random.choice(list('abcdef123456789'), 6))
```

### Plot Top 10 Job Title

```python
index = df['Job Title'].value_counts()[:11].index
values = df['Job Title'].value_counts()[:11].values

label = {"fontsize":15, "color":"grey", "fontfamily":"serif"}
title = {"fontsize":15, "color":"#6acbff", "fontfamily":"serif"}

fig, ax = plt.subplots(nrows=1, ncols=1, facecolor='black', figsize=(10, 6), dpi=100, subplot_kw=dict(facecolor="black"))


ax.spines['left'].set_color('white')
ax.spines['bottom'].set_color('white')

ax.set_xlabel("Job Title", fontdict=label)
ax.set_ylabel("Count", fontdict=label)
ax.set_title('Top 10', fontdict=title)

plt.bar(index, values, color=[colors() for c in range(len(index))])
plt.tick_params(labelcolor="white",color='white')
plt.xticks(rotation=90)

for p, val in zip(ax.patches, values):
    x, y = p.get_x()+0.2 , p.get_height()+0.2
    plt.text(x, y, val, color="white")

plt.savefig("FlaskProject/static/topjobtitle", bbox_inches="tight")
plt.show()
```
